```python
import matplotlib.pyplot as plt
  
exp_e_pos_1 = []
exp_m_pos_1 = []
for line in open('Exp_e_m_1.txt', 'r'):
    lines = [i for i in line.split()]
    exp_e_pos_1.append(float(lines[0]))
    exp_m_pos_1.append(float(lines[1]))
```
