##### Start with all spins beieng positive

Temprature T = 1

Read the data

```python
import matplotlib.pyplot as plt
  
exp_e = []
exp_m = []
for line in open('Exp_e_m_1.txt', 'r'):
    lines = [i for i in line.split()]
    exp_e.append(float(lines[0]))
    exp_m.append(float(lines[1]))
```

```python
fig = plt.figure()
plt.title('Expectation value e for cycles, start with only positiv spins, T=1', fontsize=15)
plt.xlabel('Cycle', fontsize=15)
plt.ylabel('Expectation value e', fontsize=15)
plt.xscale('symlog')
plt.plot(exp_e)
  
plt.show()
```

Make excpectation values for the energy

```python
fig = plt.figure()
plt.title('Expectation value e for cycles, start with only positiv spins, T=1', fontsize=15)
plt.xlabel('Cycle', fontsize=15)
plt.ylabel('Expectation value e', fontsize=15)
plt.xscale('symlog')
plt.plot(exp_m)
  
plt.show()
```
