Temprature T = 1

Read the data

```python
import matplotlib.pyplot as plt
  
exp_e_pos_1 = []
exp_m_pos_1 = []
for line in open('Exp_e_m_1.txt', 'r'):
    lines = [i for i in line.split()]
    exp_e_pos_1.append(float(lines[0]))
    exp_m_pos_1.append(float(lines[1]))
```

```python
fig = plt.figure(figsize=(8, 6))
plt.title('E[ε] plotted for number of cycles, initialized with all positiv spins, T=1', fontsize=15)
plt.xlabel('Cycles', fontsize=15)
plt.ylabel('E[ε]', fontsize=15)
plt.xscale('symlog')
plt.plot(exp_e_pos_1)
#plt.savefig('e_INIT_pos_T_1.pdf')
plt.show()
```

Make excpectation values for the energy

```python
fig = plt.figure(figsize=(8, 6))
plt.title('E[m] plotted for number of cycles, initialized with all positiv spins, T=1', fontsize=15)
plt.xlabel('Cycles', fontsize=15)
plt.ylabel('E[m]', fontsize=15)
plt.xscale('symlog')
plt.plot(exp_m_pos_1)
#plt.savefig('e_INIT_pos_T_1.pdf')
plt.show()
```
