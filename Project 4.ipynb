Finding the critical temperature

```python
critical40 = data40_mill.iloc[:,[3]].max()
print(critical40)
print(data40_mill.idxmax())
critical60 = data60_mill.iloc[:,[3]].max()
print(critical60)
print(data60_mill.idxmax())
crtical80 = data80_mill.iloc[:,[3]].max()
print(data80_mill.idxmax())
print(crtical80)
crtical100 = data100_mill.iloc[:,[3]].max()
print(data100_mill.idxmax())
```

```python
temp40 = data40_mill.iloc[17,:]
print(temp40[0])
temp60 = data60_mill.iloc[32,:]
print(temp60[0])
temp80 = data80_mill.iloc[16,:]
print(temp80)
temp100 = data100_mill.iloc[32,:]
print(temp100)
```

```python

lat_size = np.array([[1/40], [1/60], [1/80], [1/100]])
crit_vals = np.array([[temp40[0]], [temp60[0]], [temp80[0]], [temp100[0]]])

fig = plt.figure(figsize=(10, 8))
#plt.title('Susceptibility [X] plotted for different temperatures', fontsize=18)
plt.xlabel(r'$L^{-1}$', fontsize=19)
plt.ylabel(r'$T_c$', fontsize=19)
#plt.xscale('symlog')
plt.grid()
plt.plot(lat_size, crit_vals, '-o', markersize = 7)
plt.xticks(size = 18)
plt.yticks(size = 18)
plt.savefig('crit_T_T.pdf')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(lat_size, crit_vals, sample_weight = None)
print(model.coef_)
print(model.intercept_)
```

```python
new = np.array(np.linspace(0,1/100,10)).reshape(-1, 1)
pred = model.predict(lat_size)
```
